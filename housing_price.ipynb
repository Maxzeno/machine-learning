```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
import joblib
```

```python
# Read the dataset and remove the columns with non numeric data
df = pd.read_csv("/content/drive/MyDrive/datasets/USA_Housing.csv")
df.pop('Address')

# Get the target and the label
y = df['Price']
scaler = MinMaxScaler()
X = df.drop('Price', axis=1)
```

```python
# Plot heatmap to see correlation within the data
sns.heatmap(df.corr())
```

```python
# Split the train and test set
X_train, X_test, y_train, y_test = train_test_split(scaler.fit_transform(X.to_numpy()), y.values, test_size=0.1, random_state=42)
```

```python
# Instantiate the LinearRegression class class from sklearn and fit the train set 
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# View the weights and bias
model.coef_, model.intercept_
```

```python
# Make prediction
yhat = model.predict(X_test)

# Plot first 10 of predictions and tru value of y (y test)
plt.plot(yhat[:10])
plt.plot(y_test[:10])
plt.show()
```

```python
# Get the score of the model (check its performance with respect to accuracy)
model.score(X_test, y_test)
```

```python
# Save the model
joblib.dump(model, 'housing price.pkl')
```
